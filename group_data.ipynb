````python
for name, group_df in df.groupby(by=['species']):
    print(name)
    print(group_df.head(3))
````



###  groupby + agg

> group 으로 묶은 후 통계 함수를 사용


```
agg function list

count() – Number of non-null observations
sum() – Sum of values
mean() – Mean of values
median() – Arithmetic median of values
min() – Minimum
max() – Maximum
mode() – Mode
std() – Standard deviation
var() – Variance
```

````python
df.groupby(by=['species']).agg(['std','median','count'])
````
