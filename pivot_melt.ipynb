```python
import pandas as pd
```

```python
df=pd.DataFrame({'store':['Costco','Costco','Costco','Wal-Mart','Wal-Mart','Wal-Mart',"Sam's Club","Sam's Club","Sam's Club"],
               'product':['Potato','Onion','Cucumber','Potato','Onion','Cucumber','Potato','Onion','Cucumber'],
               'price':[3000,1600,2600,3200,1200,2100,2000,2300,3000],
                'quantity':[25,31,57,32,36,21,46,25,9]})
df
```

```python
df.set_index('store').loc['Costco', :]
```

```python
df.pivot(index='store', columns='product', values='quantity')
```

```python
pivot_df = df.pivot_table(index='store', columns='product', values='quantity')
```

```python
pivot_df.reset_index().melt(id_vars='store',value_name='quantity', value_vars=['Cucumber', 'Onion', 'Potato'])
```
