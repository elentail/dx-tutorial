## pd.DataFrame.iterrows() : 행에 대해 순환 반복
    (Iterate over DataFrame rows as (index, Series) pairs.)
    
    
### iterrows



```
Signature: df.iterrows() -> 'Iterable[tuple[Hashable, Series]]'
Docstring:
Iterate over DataFrame rows as (index, Series) pairs.
```


**더 자세히 풀어 쓰면**

```
Iterable[tuple[Hashable, Series]] > Key 와 행으로 이루어진 List 를 반환 

[ (key1, row1), (key2, row2), (key3, row3), ...  ]

```

___

**sereis 란**



````python
for key,row in df.iterrows():
    
    # series 혹 dataframe 을 list로 바꾸는 방법은 .values 를 호출하면 됩니다.
    
    print(key, row)
    # print(key, row.values)
    # print(key, row.sepal_length, row['sepal_length'])
````

## pd.DataFrame.itertuples() : 이름이 있는 튜플 (인덱스, 행, 열) 에 대해 순환 반복
    (Iterate over DataFrame rows as namedtuples)

### itertuples

```
Signature: df.itertuples(index: 'bool' = True, name: 'str | None' = 'Pandas') -> 'Iterable[tuple[Any, ...]]'
Docstring:
Iterate over DataFrame rows as namedtuples.

```

**더 자세히 풀어 쓰면**

```
Iterable[tuple[Any, ...]] > 행으로 이루어진 List 를 반환 

[ (row1), (row2), (row3), ...  ]

```

**iterrows 와 차이점**

- iterrows > (Key, Series) 반환
- itertuples > tuple 을 반환

````python
for row in df.itertuples():
    print(row)
    # row['Index'] [ X ] > row.Index [ O ]
    # row['sepal_length'] [ X ] > row.sepal_length [ O ]
    # print(row.Index, row.sepal_length, row.petal_length)
````

## apply 를 통해 하나의 컬럼값 연산하기
    (apply 는 pandas 객체에 열 혹은 행에 대해 함수를 적용하게 해주는 함수 )
    
##  apply 를 통해 하나의 컬럼값 연산 후 새로운 컬럼에 재할당


## apply 를 통해 여러개의 컬럼 연산하여 하나의 컬럼값 도출
    

````python
df
````

### loc 행 순회

````python
for key in df.index:
    # print(df.loc[key, :])
    print(df.loc[key, :].values)
````

### loc 열 순회

````python
for col_name in df.columns:
    # print(df.loc[key, :])
    print(col_name, df.loc[:, col_name].values)
````

## apply

### apply 를 통해 하나의 컬럼값 연산하기

````python
# 함수
def clf(x):
    if x > 6:
        return 1
    else:
        return 0
    
df['sepal_length'].apply(clf)
````

### apply 를 통해 하나의 컬럼값 연산하기(lambda version)

````python
df['sepal_length'].apply(lambda x : 1 if x > 6 else 0)
````

### apply 를 통해 하나의 컬럼값 연산 후 새로운 컬럼에 재할당

````python
# 함수
def clf(x):
    if x > 6:
        return 1
    else:
        return 0
    
df['clf_sepal_length'] = df['sepal_length'].apply(clf)
df
````

### apply 를 통해 여러개의 컬럼 연산하여 하나의 컬럼값 도출

```
Signature:
df.apply(
    func: 'AggFuncType',
    axis: 'Axis' = 0,
    raw: 'bool' = False,
    result_type: "Literal['expand', 'reduce', 'broadcast'] | None" = None,
    args=(),
    **kwargs,
)
Docstring:
Apply a function along an axis of the DataFrame.

Parameters
----------
func : function
    Function to apply to each column or row.
axis : {0 or 'index', 1 or 'columns'}, default 0
    Axis along which the function is applied:

    * 0 or 'index': apply function to each column.
    * 1 or 'columns': apply function to each row.
```



**컬럼 한개를 선택하면 문제가 되지 않지만, DataFrame 자체를 선택하면 항상 axis 확인 해야됨**
```

axis : {0 or 'index', 1 or 'columns'}, default 0
    Axis along which the function is applied:

    * 0 or 'index': apply function to each column.
    * 1 or 'columns': apply function to each row.
```

````python
# 함수
def clf_multiple(x):
    if x['sepal_length'] + x['sepal_width'] > 9:
        return 1
    
    else:
        return 0
    
df['clf_multi_sepal_length'] = df.apply(clf_multiple, axis=1)
df
````
